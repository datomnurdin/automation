```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas
from pandas import DataFrame
import csv
```

```python
with open('data.csv', mode='w') as csv_file:
    fieldnames = ['link', 
                  'name', 
                  'set',
                  'image',
                  'seller',
                  'condition', 
                  'price']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
```

```python
card_link = []
card_name = []
card_set = []
card_image = []
card_seller = []
card_condition = []
card_price = []
```

```python
def scraping(webpage, page_number):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}
    next_page = webpage + str(page_number)
    response= requests.get(str(next_page), headers=headers)
    soup = BeautifulSoup(response.text,"html.parser")
    
    soup_content = soup.findAll("div", { "class": "card h-100 p-3"})
    
    for x in range(len(soup_content)):
        link = soup_content[x].find('a', { "class": "card-text" })['href']
        #print(link)
        card_link.append(link)
        
        name = soup_content[x].find('a', { "class": "card-text" }).text
        #print(name)
        card_name.append(name)
        
        set = soup_content[x].find('div', { "class": "col-12 prod-cat" }).find('a').text
        #print(set)
        card_set.append(set)
        
        image = soup_content[x].find('img', { "class": "productImage w-100 px-2 pt-1 pb-3 px-sm-3" })['src']
        #print(image)
        card_image.append(image)
        
        seller = soup_content[x].findAll('div', { "class": "row position-relative align-center py-2 m-auto" })[0].findAll("div", { "class": "col-3 text-center p-1"})[0].find("img")["title"]
        #print(seller)
        card_seller.append(seller)
        
        condition = soup_content[x].findAll('div', { "class": "row position-relative align-center py-2 m-auto" })[0].findAll("div", { "class": "col-3 text-center p-1"})[1].text.strip()
        #print(condition)
        card_condition.append(condition)
        
        price = soup_content[x].findAll('div', { "class": "row position-relative align-center py-2 m-auto" })[0].findAll("div", { "class": "col-2 text-center p-1"})[0].text.strip()
        #print(price)
        card_price.append(price)

    if page_number < 6:
        page_number = page_number + 1
        scraping(webpage, page_number)
```

```python
scraping('https://www.trollandtoad.com/category.php?min-price=&max-price=&items-pp=60&item-condition=&search-words=hyper+rare&selected-cat=7061&sort-order=H-L&view=list&subproduct=0&page-no=', 1)
```

```python
data = { 
    'link': card_link,
    'name':card_name, 
    'set': card_set, 
    'seller': card_seller,
    'image': card_image,
    'condition': card_condition,
    'price': card_price
}
```

```python
df = DataFrame(data, columns = ['link', 
                  'name', 
                  'set', 
                  'seller', 
                  'image',
                  'condition', 
                  'price'])
df.to_csv('data.csv')
```
